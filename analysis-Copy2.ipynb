```python
import pandas as pd
import numpy as np
```

```python
# read the excel sheet
filename = ''
filename2 = ''
df = pd.read_excel(filename, 'Sheet1')
df2 = pd.read_excel(filename2, 'Sheet2')
```

```python
df2.head()
```

### Ratio vs spacer length when the tail is 37 aa and signal is Bar

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # use seaborn styles

sns.set_style('ticks',
              {'xtick.direction': u'in',
               'ytick.direction': u'in',
               'text.color': 'black',
               'text.size': '14'})

df2[
    (df2['Signal'] == 'Bar') & ((df2['Tail_Length'] == 37) & ~(df2['Construct_Terminus'] == 'ODC_var'))
    ].pivot_table('Ratio', 
                  index=['Spacer_Length'], 
                  columns=['Construct_Terminus']
        ).plot()

plt.ylabel('YFP/RFP',
           color='black',
           fontsize='16')

plt.xlabel('Spacer Length',
           color='black',
           fontsize='16')

plt.title('Constructs with N-Terminal Bar, Variable Spacer, \nand Last 37 aa of Tail (ie ODC_var excluded)', 
          color='black', fontsize='18')
```

### Ratio vs spacer length when the tail is 37 aa and signal is UbL

```python
%matplotlib inline
sns.set() # use seaborn styles

sns.set_style('ticks',
              {'xtick.direction': u'in',
               'ytick.direction': u'in',
               'text.color': 'black',
               'text.size': '14'})

df2[
    (df2['Signal'] == 'UbL') & ((df2['Tail_Length'] == 37) & ~(df2['Construct_Terminus'] == 'ODC_var'))
    ].pivot_table('Ratio',
                  index=['Spacer_Length'],
                  columns=['Construct_Terminus']
                 ).plot()

plt.ylabel('YFP/RFP',
           color='black',
           fontsize='16')

plt.xlabel('Spacer Length',
           color='black',
           fontsize='16')

plt.title('Constructs with N-Terminal UbL, Variable Spacer,\nand Last 37 aa of Tail (ie ODC_var excluded)', 
          color='black',
          fontsize='18')
```

### Ratio vs spacer length when the tail is 37 aa and signal is UbL (SP2 excluded)

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

sns.set() # use seaborn styles
sns.set_style('ticks',
              {'xtick.direction': u'in',
               'ytick.direction': u'in',
               'text.color': 'black',
               'text.size': '14'})
sns.despine(left=True, bottom=True, right=False)

df2[
    (df2['Signal'] == 'UbL') & ((df2['Tail_Length'] == 37) & ~(df2['Construct_Terminus'] == 'ODC_var') & ~(df2['Construct_Terminus'] == 'spacer_SP2'))
    ].pivot_table('Ratio',
                  index=['Spacer_Length'],
                  columns=['Construct_Terminus']
                 ).plot()

plt.ylabel('YFP/RFP',
           color='black',
           fontsize='16')

plt.xlabel('Spacer Length',
           color='black',
           fontsize='16')

plt.title('Constructs with N-Terminal UbL, Variable Spacer, \nand Last 37 aa of Tail (spacer_SP2 and ODC_var excluded)',
          color='black',
          fontsize='18')
```

### Ratio vs total initiation region length when the signal is Bar
"ODC_var" indicates that the initiation region in question varies by the length of ODC C-terminus section, rather than by spacer before terminal section

```python
%matplotlib inline

plt.figure(num=1, dpi=100)
sns.set() # use seaborn styles

sns.set_style('ticks',
              {'xtick.direction': u'in',
               'ytick.direction': u'in',
               'text.color': 'black',
               'text.size': '14'})


df2[(df2['Signal'] == 'Bar') & ((df2['Tail_Length'] == 37) | (df2['Construct_Terminus'] == 'ODC_var'))
    ].pivot_table('Ratio',
                  index=['Total_Length'],
                  columns=['Construct_Terminus']
                 ).plot()

plt.ylabel('YFP/RFP',
           color='black',
           fontsize='16')

plt.xlabel('Total Initiation Region Length',
           color='black',
           fontsize='16')

plt.title('Constructs with N-terminal Bar',
          color='black',
          fontsize='18')
```

### Ratio vs total initiation region length when the signal is UbL 

```python
%matplotlib inline

sns.set() # use seaborn styles
sns.set_style('ticks',
              {'xtick.direction': u'in',
               'ytick.direction': u'in',
               'text.color': 'black',
               'text.size': '14'})

df2[(df2['Signal'] == 'UbL') & ((df2['Tail_Length'] == 37) | (df2['Construct_Terminus'] == 'ODC_var'))
    ].pivot_table('Ratio',
                  index=['Total_Length'],
                  columns=['Construct_Terminus']
                 ).plot(antialiased=True)

plt.ylabel('YFP/RFP',
           color='black',
           fontsize='16')

plt.xlabel('Total Initiation Region Length',
           color='black',
           fontsize='16')

plt.title('Constructs with N-terminal UbL',
          color='black',
          fontsize='18')
```
